```python
from nltk.corpus import inaugural
import string
import operator
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
x=[]
stop_words = set(stopwords.words('english'))
k=inaugural.words('2009-Obama.txt')
#k=[s.translate(None,string.punctuation) for s in k]
k=[''.join(c for c in s if c not in string.punctuation) for s in k]
for w in k:
    if w not in stop_words: 
        x.append(w)

```

```python
l={}
new=[]
k={}
z=set(x)

for word in z:
 
    l[word]=x.count(word)
#print(l)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for words in x:
    new.append(ps.stem(words))
p=set(new)
for w in p:
    k[w]=new.count(w)
plt.plot(k.values())
#plt.xlabel(k.keys())
k_sorted = sorted(k.items(), key=operator.itemgetter(1),reverse=True)
del(k_sorted[0])
print(k_sorted)
```
